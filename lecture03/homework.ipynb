```python
import numpy as np
from numpy.linalg import svd

m = 128
n = 128
A = np.random.randn(m,n)
x = np.zeros((n,1))
b = np.random.randn(m,1)

U,S,Vh = svd(A)
D = np.diag(S**(-1))
sol = Vh.T @ D @ U.T @ b
```

```python
def calculate_norm(x,y,n):
    norm = 0
    for idx in range(n):
        norm += (x[idx] - y[idx])**2 / (abs(y[idx]) + 1e-8)**2
    
    norm /= n
    norm = norm ** 0.5
    return norm

relative_tol = calculate_norm(A@sol, b, m)
print(relative_tol)
```
